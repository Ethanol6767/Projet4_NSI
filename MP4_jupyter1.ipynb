## <u>Exercice 1 :</u>

Une image est une reproduction visuelle d'un objet réel. Une image numérique est une image (dessin, icône, photographie...) créée, traitée, stockée sous forme binaire (suite de 0 et de 1). Elle est composée d'un ensemble d'éléments de base appelés pixels.
1. La définition est le nombre de points (ou pixels) que comporte une image numérique en largeur et en hauteur (le nombre de colonnes et nombre de lignes). On parle aussi de Taille en pixels.
2. La résolution d'une image est le nombre de pixels contenus dans l'image par unité de longueur. Elle s'exprime le plus souvent en ppp (pixels par pouces) ou en dpi (dots per inch), parfois en pixels par cm. La résolution définit la netteté et la qualité d'une image. Plus la résolution est grande (c'est-à-dire plus il y a de pixels dans une longueur de 1 pouce), plus votre image est précise dans les détails.
3. La profondeur définit la quantité d’informations chromatiques disponibles pour chaque pixel dans une image. Le nombre de couleurs disponibles et la précision de la représentation des couleurs dans une image sont proportionnels au nombre de bits d’informations par pixel. Ainsi, dans une image d’une profondeur de 1 bit par pixel, les pixels peuvent prendre deux valeurs possibles : noir et blanc. Une image avec une profondeur de 8 bits par pixel compte 256 valeurs possibles (28). Une image en mode Niveaux de gris avec une profondeur de 8 bits par pixel compte 256 valeurs possibles de gris.

Soit x et y les nombres de pixels minimals respectivement en longueur et en hauteur d'une image pour obtenir une résolution de 300ppp dans le cas d’une impression sur du papier photo de dimensions 10 cm x 15 cm.

2,54cm = 1pouce<br/>
10cm ≈ 3,937pouces<br/>
15cm ≈ 5,906pouces

$$Résolution = \frac{x}{3,937} = 300ppp$$

$$x = 300 \times 3,937 = 1181,1 ≈ 1181px$$

$$Résolution = \frac{y}{5,906} = 300ppp$$

$$y = 300 \times 5,906 = 1771,8 ≈ 1772px$$

La définition minimale correspond donc plus ou moins à $1181\times1772$ pixels
